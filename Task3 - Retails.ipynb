## Exploratory Data Analysis

> **1. Calculating the profit of each Category and Subcategory**

```python
plt.figure(figsize=(9,5), dpi = 100)

#On x-axis 
plt.xlabel('Category', fontsize = 15)
#On y-axis 
plt.ylabel('Profit', fontsize=15)
#Name of the graph
plt.title('Profit of each Category', fontsize=15)

#giving a histogram plot
plt.hist(Retails['Category'], rwidth = 1, bins =15)
#displays the plot
plt.show()
```

> So the Office Supplies had the most and Tecgnology had the least Profit.

```python
plt.figure(figsize=(9,5), dpi = 100)

#On x-axis 
plt.xlabel('Sub-Category', fontsize = 15)
plt.xticks(rotation=45)
#On y-axis 
plt.ylabel('Profit', fontsize=15)
#Name of the graph
plt.title('Profit of each Sub-Category', fontsize=15)

#giving a histogram plot
plt.hist(Retails['Sub-Category'], rwidth = 0.9, bins =35)
#displays the plot
plt.show()
```

> So Blinders had most and Copiers had least!

> **2. Calculating the profit of each State**

```python
plt.figure(figsize=(9,5), dpi = 100)

#On x-axis 
plt.xlabel('State', fontsize = 15)
plt.xticks(rotation=90)
#On y-axis 
plt.ylabel('Profit', fontsize=15)
#Name of the graph
plt.title('Profit from each State', fontsize=15)

#giving a histogram plot
plt.hist(Retails['State'], rwidth = 0.9, bins =35)
#displays the plot
plt.show()
```

> So Kentucky had the most and West Virginia had least Profit.

> **3. Calculating the Quantity of each Sub-Category**

```python
plt.figure(figsize=(9,5), dpi = 100)

#On x-axis 
plt.xlabel('Sub-Category', fontsize = 15)
plt.xticks(rotation=45)
#On y-axis 
plt.ylabel('Quantity', fontsize=15)
#Name of the graph
plt.title('Purchased Quantity of each Sub-Category', fontsize=15)

#giving a histogram plot
plt.hist(Retails['Sub-Category'], rwidth = 0.9, bins =35)
#displays the plot
plt.show()
```

> As expected the most sub-category had profit is the most purchased, and the laest had profit is least purchased.(Blinders,Copiers)!

> **4. Calculating Sales for each Region**


```python
plt.figure(figsize=(9,5), dpi = 75)

#On x-axis 
plt.xlabel('Region', fontsize = 15)
plt.xticks(rotation=0)
#On y-axis 
plt.ylabel('Sales', fontsize=15)
#Name of the graph
plt.title('Sales Amount for each Region', fontsize=15)

#giving a histogram plot
plt.hist(Retails['Region'], rwidth = 0.9, bins =25)
#displays the plot
plt.show()
```

> West has most, and South had least in general.

> **5. Calculating Purchased amount for each City**


```python
plt.figure(figsize=(9,5), dpi = 100)

#On x-axis 
plt.xlabel('City', fontsize = 15)
plt.xticks(rotation=135)
#On y-axis 
plt.ylabel('Quantity', fontsize=15)
#Name of the graph
plt.title('Purchased Amount for each City', fontsize=15)

#giving a histogram plot
plt.hist(Retails['City'], rwidth = 0.9, bins =45)
#displays the plot
plt.show()
```

> Distributions of Purchased amount in each city.
